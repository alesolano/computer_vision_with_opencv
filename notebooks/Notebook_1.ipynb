#### 1. Load the images ‘Connector01.jpg’ and  ‘Cables14.jpg’ and show them in two different figures (im1 and im2). 

To convert an image from a compressed file extension (JPEG, PNG...) to a multidimensional (`numpy`) array, we use `cv2.imread()`. This array will have dimensions `(height, width)` for grayscale images and `(height, width, 3)` for RGB images.

It's important to set the second field of `cv2.imread()` to `0` for grayscale images. This way, `OpenCv` creates an array with shape `(height, width)`. You should also set the second field of `cv2.imshow()` to `gray` in this case.

A very annoying feature of `OpenCV` is that it loads images into BGR format. However, `matplotlib` works with RGB format. This means we need to convert images to RGB each time we load them. 

```python
im1 = cv2.imread('../Images/Connector/Connector 01.jpg', 0) # grayscale image
im2_bgr = cv2.imread('../Images/cables/cables14.jpg') # bgr image
im2 = cv2.cvtColor(im2_bgr, cv2.COLOR_BGR2RGB) # rgb image

plt.figure(0)
plt.imshow(im1, 'gray')

plt.figure(1)
plt.imshow(im2)

plt.show()
```

#### 2. Visualize both images (im1 and im2) in a single figure.

```python
plt.figure(2)
plt.subplot(1,2,1), plt.imshow(im1, 'gray')
plt.subplot(1,2,2), plt.imshow(im2)
plt.show()

# Here we see the shapes of both images
print(im1.shape)
print(im2.shape)
```

#### 4. Show in a new figure the row 185 of the image `im1`, and enlight that row in green color upon the image.

`[185]*im1.shape[1]` is a list with value 185 and shape equals to the width of the image `im1`. This draws the green line upon the image.

```python
plt.figure(3)
plt.subplot(2,1,1), plt.plot(im1[185,:], 'g')
plt.subplot(2,1,2), plt.imshow(im1, 'gray'), plt.plot([185]*im1.shape[1], 'g')
plt.show()
```

#### 5. Show the image `im1` with full resolution, with ½ resolution, with ¼ resolution and with 1/8 resolution.

In each loop iteration the image resolution is divided by a power of 2.

`cv2.resize()` has an attribute `interpolation` which can be `cv2.INTER_AREA`, `cv2.INTER_CUBIC` and `cv2.INTER_LINEAR` (value by default).

```python
plt.figure(4)

for i in range(4):
    new_heigth = round(im1.shape[0]/(2**i))
    new_width = round(im1.shape[1]/(2**i))
    im1_resized = cv2.resize(im1, (new_width, new_heigth), interpolation=cv2.INTER_CUBIC)
    plt.subplot(2,2,(i+1)), plt.imshow(im1_resized, 'gray')
    
plt.show()
```

#### 6. Show the image `im2` along with its three fields: R, G and B.

0 -> red
1 -> green
2 -> blue

```python
im2_R = im2[:,:,0];
im2_G = im2[:,:,1];
im2_B = im2[:,:,2];

plt.figure(5)
plt.subplot(2,3,2), plt.imshow(im2)
plt.subplot(2,3,4), plt.imshow(im2_R, 'gray')
plt.subplot(2,3,5), plt.imshow(im2_G, 'gray')
plt.subplot(2,3,6), plt.imshow(im2_B, 'gray')
plt.show()
```

#### 7. Convert the image `im2` from RGB into HSV format. Show the original image `im2` and its H, S, V components. 

```python
im2_HSV = cv2.cvtColor(im2, cv2.COLOR_RGB2HSV)

im2_H = im2_HSV[:,:,0];
im2_S = im2_HSV[:,:,1];
im2_V = im2_HSV[:,:,2];

plt.figure(6)
plt.subplot(2,3,2), plt.imshow(im2)
plt.subplot(2,3,4), plt.imshow(im2_H, 'gray')
plt.subplot(2,3,5), plt.imshow(im2_S, 'gray')
plt.subplot(2,3,6), plt.imshow(im2_V, 'gray')
plt.show()
```
